### Preprocessing labels a One hot encoded 

```python
unique_labels = np.unique(labels)
labels_num_dict = {unique_labels[index]:index for index in range(len(unique_labels))}
num_labels_dict = {labels_num_dict[key]:key for key in list(labels_num_dict.keys())}
new_labels = np.array([labels_num_dict[label] for label in labels])
new_labels = np.eye(len(set(new_labels)))[new_labels]
new_labels
```

## Modelo implementado

```python
class KNN():
    def __init__(self):
        self.labels_num_dict = {}
        self.num_labels_dict = {}
    
    def create_dictionaries(self, labels):
        unique_labels = np.unique(labels)
        self.labels_num_dict = {unique_labels[index]:index for index in range(len(unique_labels))}
        self.num_labels_dict = {labels_num_dict[key]:key for key in list(labels_num_dict.keys())}
    
    
    def traducir_num_to_string(self, num_labels):
        return [self.num_labels_dict[num_label] for num_label in num_labels]

    def predecir_raza_KNN(self, x_predict, X_train, Y_train, k):
        
        x_predict = x_predict.reshape((1, -1))
        self.create_dictionaries(Y_train)
        
        g = tf.Graph()
        with g.as_default():
            x_data_train = tf.placeholder(shape=X_train.shape, dtype=tf.float32)
            y_data_train = tf.placeholder(shape=Y_train.shape, dtype=tf.float32)
            x_data_test = tf.placeholder(shape=x_predict.shape, dtype=tf.float32)

            distance = tf.sqrt(tf.reduce_sum(tf.square(tf.subtract(x_data_train, tf.expand_dims(x_data_test, 1))), axis=2))
            _, top_k_indices = tf.nn.top_k(tf.negative(distance), k=k)
            top_k_label = tf.gather(y_data_train, top_k_indices)

            sum_up_predictions = tf.reduce_sum(top_k_label, axis=1)
            prediction = tf.argmax(sum_up_predictions, axis=1)


            with tf.Session(graph=g) as session:
                tf.global_variables_initializer().run()
                writer = tf.summary.FileWriter(f'./knn_tensorboard/KNN', session.graph)
                prediction_outcome = session.run(prediction, feed_dict={x_data_train: X_train, x_data_test: x_predict, y_data_train: Y_train})
                return self.traducir_num_to_string(prediction_outcome)

    
modelo = KNN()
modelo.predecir_raza_KNN(features[15], features, new_labels, 10)
```
