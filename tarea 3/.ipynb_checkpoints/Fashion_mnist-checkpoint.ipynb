### Formatting the dataset

```python
train_images, train_labels = train_images.reshape((-1, 28 * 28)).astype(np.float32), (np.arange(3) == train_labels[:,None]).astype(np.float32)
test_images, test_labels = test_images.reshape((-1, 28 * 28)).astype(np.float32), (np.arange(3) == test_labels[:,None]).astype(np.float32)

print('Training set', train_images.shape, train_labels.shape)
print('Test set', test_images.shape, test_labels.shape)
```

### Aplicar regresión logística con entropía cruzada como función de costo, crear una función que recibe de entrada un tensor x y predice la clase(utilizar el diccionario para mostrar el texto perteneciente al label).

```python
class LogisticRegressionMultiClass:
    
    def parse_config(self, config_string):
        parameters = {}
        for conf in config_string.split('_')[1:]:
            parameters[conf.split('=')[0]] = conf.split('=')[1]
        return parameters
    
    def mini_batches(self, x, y, batch_size):
        for batch_iteration in range(int(len(x)/batch_size)):
            init_batch = batch_iteration * batch_size
            end_batch = init_batch + batch_size
            yield np.array(x[init_batch:end_batch]), np.array(y[init_batch:end_batch])
            
    def accuracy(self, predictions, labels):
        return (100.0 * np.sum(np.argmax(predictions, 1) == np.argmax(labels, 1))
          / predictions.shape[0])
    
    def train(self, x, y, conf_string):
        config = self.parse_config(conf_string)        
        lr, epochs, batch_size, log_epoch = float(config['lr']), int(config['epochs']), int(config['batchsize']), int(config['logepoch'])
        
        graph = tf.Graph()
        with graph.as_default():
            x_train = tf.placeholder("float32", shape=(batch_size, 28 * 28))
            y_train = tf.placeholder("float32", shape=(batch_size, 3))
            x_test = tf.constant(test_images)

            weights = tf.Variable(tf.truncated_normal([28 * 28, 3]))
            biases = tf.Variable(tf.zeros([3]))

            logits = tf.matmul(x_train, weights) + biases
            
            loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=y_train))
            optimizer = tf.train.GradientDescentOptimizer(lr).minimize(loss)

            train_prediction = tf.nn.softmax(logits)
            test_prediction = tf.nn.softmax(tf.matmul(x_test, weights) + biases)
        
            with tf.Session(graph=graph) as session:
                writer = tf.summary.FileWriter(f'./experimentos_fashion_mnist/{conf_string}', session.graph)
                tf.global_variables_initializer().run()
                step = 1
                for epoch in range(epochs):
                    for mini_x, mini_y in self.mini_batches(x, y, batch_size):
                        feed_dict = {x_train: mini_x, y_train: mini_y} 
                        _, error, predictions = session.run([optimizer, loss, train_prediction], feed_dict=feed_dict)
                        if (step % (log_epoch * int(len(train_images)/batch_size)) == 0):
                            print("Epoch: %d, error: %f" % (epoch, error))
                            print("Minibatch accuracy: %.1f%%" % self.accuracy(predictions, mini_y))
                            print("Validation accuracy: %.1f%%" % self.accuracy(test_prediction.eval(), test_labels))
        #                     writer.add_summary(session.run(l, feed_dict=feed_dict), epoch)
                        step += 1
 
        
        
    
logistic = LogisticRegressionMultiClass()
logistic.train(train_images, train_labels, 'logistic_epochs=100_lr=0.001_batchsize=256_logepoch=10')
```
